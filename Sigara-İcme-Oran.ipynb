```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
smoke=pd.read_csv('Desktop/smoking.csv')
```

```python
smoke.head(10)
```

```python
smoke.info()
```

```python
clean_data=smoke.dropna()
print(clean_data)
```

```python
cleaned_data=smoke.dropna(axis=1)
print(cleaned_data)
```

```python
smoke['amt_weekends'].fillna(6,inplace=True)
smoke['amt_weekdays'].fillna(smoke['amt_weekdays'].mean(),inplace=True)
print(smoke)
```

```python
smoke.drop('type',axis=1,inplace=True)
print(smoke)
```

```python
#Temizlenen veriler ve atamalar sonrası son bakış
smoke.head(10)
```

```python
selected_rows = smoke[(smoke['age'] > 35) & (smoke['marital_status'] == 'Divorced') & (smoke['smoke'] == 'Yes')]

#Divorced olanları alalım
divorced = smoke[smoke['marital_status'] == 'Divorced']

# Histogram grafiğini oluşturalım
plt.hist(divorced['age'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of Divorced Individuals')
plt.show()

#Bu kod, 'marital_status' sütunu 'Divorced' olan kişilerin yaş dağılımını gösteren bir histogram grafiği oluşturur
```

```python
# 'Divorced' olan kişilerin sayısını bulalım

divorced_count = (smoke['marital_status'] == 'Divorced').sum()

# 'divorced_count' ile 'amt_weekends' arasındaki korelasyonu hesaplayalım
correlation = smoke['age'].corr(smoke['amt_weekends'])

print("Divorced Count:", divorced_count)
print("Korelasyon:", correlation)
```
